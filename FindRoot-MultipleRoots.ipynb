```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def FindRoot(f, a, b, tol):
    left = a
    right = b
    while right-left > tol:
        midpoint = (left+right)/2
        if(f(midpoint) < 0):
            left = midpoint
        else:
            right = midpoint
    return midpoint
```

```python
def f(x):
    return np.sin(x)
```

```python
x = np.arange(-2,8,0.1)
vf = np.vectorize(f)
y = vf(x)
```

```python
plt.grid(True)
plt.plot(x,y)
```

```python
FindRoot(f, -1 , 7.2, 1e-6)
```

```python
FindRoot(f, -1 , 7.3, 1e-6)
```

```python
def g(x):
    if x < 0:
        return x-2
    else:
        return 2-x
```

```python
x = np.arange(-1,1,0.005)
vg = np.vectorize(g)
y = vg(x)
```

```python
plt.grid(True)
plt.plot(x,y)
```

```python
FindRoot(g, -1.2, 1.34, 1e-9)
```
